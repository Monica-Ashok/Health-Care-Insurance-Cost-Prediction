From the above observations and the nature of dataset, we might want to treat outliers in bmi, average steps.

```python
def remove_outlier(col):
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lr= Q1-(1.5 * IQR)
    ur= Q3+(1.5 * IQR)
    return lr, ur
```

```python
for i in numeric_cols:
    LR, UR = remove_outlier(df[i])
    df[i] = np.where(df[i] > UR, UR, df[i])
    df[i] = np.where(df[i] < LR, LR, df[i])
```

```python
plt.subplots(figsize=(10,12))

for i, col in enumerate(numeric_cols):
    plt.subplot(4, 2, i + 1)
    sns.boxplot(y=df[col], color='cyan')
    plt.title(f'{col}')
    plt.tight_layout()
```

Okay, now that we have detected and treated the missing values and outliers let us look into the column "cholesterol_level"

Cholestrol has values in the range like 120 to 150. For the easiness and accuracy of the model let's consider the midpoint of the range

```python
# define a function to extract the values from the range and calculate the midpoint
def calculate_midpoint(value):
    range_values = value.split(' to ')
    midpoint = (int(range_values[0]) + int(range_values[1])) / 2
    return midpoint

# apply the function to the 'cholesterol_level' column using the apply method
df['cholesterol_level'] = df['cholesterol_level'].apply(lambda x: calculate_midpoint(x))
```

**Interpretation of R-squared**

The R-squared value tells us that our model can explain 94% of the variance in the training set.

**Interpretation of Coefficients**

The coefficients tell us how one unit change in X can affect Y.

The sign of the coefficient indicates if the relationship is positive or negative.

Multicollinearity occurs when predictor variables in a regression model are correlated. This correlation is a problem because predictor variables should be independent. If the collinearity between variables is high, we might not be able to trust the p-values to identify independent variables that are statistically significant.

When we have multicollinearity in the linear model, the coefficients that the model suggests are unreliable.

**Interpretation of p-values (P > |t|)**

For each predictor variable there is a null hypothesis and alternate hypothesis.

- Null hypothesis : Predictor variable is not significant 
- Alternate hypothesis : Predictor variable is significant

(P > |t|) gives the p-value for each predictor variable to check the null hypothesis.

If the level of significance is set to 5% (0.05), the p-values greater than 0.05 would indicate that the corresponding predictor variables are not significant.

However, due to the presence of multicollinearity in our data, the p-values will also change.

We need to ensure that there is no multicollinearity in order to interpret the p-values.

**Multicollinearity**

- If VIF is 1, then there is no correlation among the $k$th predictor and the remaining predictor variables, and  hence, the variance of $\beta_k$ is not inflated at all.
- If VIF exceeds 5, we say there is moderate VIF, and if it is 10 or exceeding 10, it shows signs of high multi-collinearity.
- The purpose of the analysis should dictate which threshold to use.

```python
# let's check the VIF of the predictors
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_series1 = pd.Series(
    [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])],
    index=X_train.columns,
)
print("VIF values: \n\n{}\n".format(vif_series1))
```

Multicollinearity affects only the specific independent variables that are 
correlated. Therefore, in this case, we can trust the p-values of last_year_admitted, regular_checkup_last_year, weight_change_last_year , weight and participated_adventure_sports as these had connection with insurance cost while we did EDA

To treat multicollinearity, we will have to drop one or more of the correlated features

#Model Tuning

After the building above models and comparing the metrics, we can conclude that Gradient Boosting Regressor is the optimal model

Now, lets try to tune and cross validate the model to see if the tuning can stabilize the model and prevent overfitting to greater extend compared to its current performance

```python
#Model Tuning Using Grid Search CV
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of boosting stages
    'learning_rate': [0.1, 0.01, 0.001],  # Step size shrinkage
    'max_depth': [3, 5, 7]  # Maximum depth of each tree
}

# Perform grid search
grid_search = GridSearchCV(estimator=gb, param_grid=param_grid, cv=5)
grid_search.fit(X_train, np.ravel(Y_train))

# Get the best parameters and the best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

```python
#Cross Validation for above models
from sklearn.model_selection import cross_val_score

# Perform cross-validation on the untuned model
untuned_scores = cross_val_score(gb, X_train, np.ravel(Y_train), cv=5, scoring='r2')

# Perform cross-validation on the tuned model
tuned_scores_grid = cross_val_score(best_model, X_train, np.ravel(Y_train), cv=5, scoring='r2')

# Print the mean R^2 scores for both models
print("Untuned Model Score:", untuned_scores.mean())
print("Grid Search Tuned Model Score:", tuned_scores_grid.mean())
```
